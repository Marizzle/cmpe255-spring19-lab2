```python
import math
def normal_pdf(x, mu=0, sigma=1):
    one_over_sqrt_2_pi_sigma_squared = 1/(math.sqrt(2*3.14*sigma*sigma))
    exp = math.exp(-(x-mu)*(x-mu)/(2*sigma*sigma))
    return one_over_sqrt_2_pi_sigma_squared*exp
```

```python
from matplotlib import pyplot as plt
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_pdf(x, sigma=1) for x in xs], '-', label='mu=0,sigma=1')
plt.plot(xs, [normal_pdf(x, sigma=2) for x in xs], '_', label='mu=0,sigma=2')
plt.plot(xs, [normal_pdf(x, sigma=0.5) for x in xs], '-', label='mu=0,sigma=0.5')
plt.plot(xs, [normal_pdf(x, sigma=-1) for x in xs], '-', label='mu=0,sigma=-1')
plt.legend()
plt.show()
```
